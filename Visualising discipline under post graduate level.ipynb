```python
# for PG level

for i,row in disc.iterrows():
    count=0
    for index,row in course.iterrows():
        val=course.iloc[index]['level_id']
#     print(val)
    
        if val==3:
        
            if course.iloc[index]['broad_discipline_group_category_id']==i+1:
                count+=1
    disc.ix[i,'freq_PG']=count
    print(count)
```
