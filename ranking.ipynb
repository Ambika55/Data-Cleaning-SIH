```python
col=[
    ('no_of_libraries',2),
    ('no_of_laboratories',4),
    ('no_of_conference_halls',1),
    ('no_of_health_centers',3),
    ('no_of_computer_centers',5)]
```

```python
print(col)
```

```python
# infra[infra['id']==college.infrastructure_id][[x[0] for x in col]].values
[x[0] for x in col]
```

```python
for index, row in college.iterrows():
    id=college.loc[index]['infrastructure_id']
    
    infra_values=infra[infra['id']==id][[x[0] for x in col]].values
    if len(infra_values)>0:
        factor = 0
        for i,weight in zip(infra_values[0],[x[1] for x in col]):
            factor+=(i*weight)
        hostel=college.iloc[index]['no_of_student_hostel'] 
        if factor+hostel is not 0:
            college.ix[index,'ranking']=factor+hostel
        else:
            college.ix[index,'ranking']=1
    else:
        college.ix[index,'ranking']=1
    college.ix[index,'PTR ratio']=np.random.randint(9,12)
```
