```python
import pandas as pd
import numpy as np
```

```python
college=pd.read_csv("./updated_college.csv")
```

```python
college.head()
```

```python
college=college.drop(columns=['city','latitude','longitude'])
```

```python
college.head()
```

```python
infra=pd.read_csv("./infrastructure.csv")
```

```python
college=college.drop(columns=['Unnamed: 0'])
```

```python
college.head()
```

```python
 col=[
    ('no_of_libraries',2),
    ('no_of_laboratories',4),
    ('no_of_conference_halls',1),
    ('no_of_health_centers',3),
    ('no_of_computer_centers',5)]
```

```python
college['Rank']=0
```

```python
for index, row in college.iterrows():
    id=college.loc[index]['infrastructure_id']
    
    infra_values=infra[infra['id']==id][[x[0] for x in col]].values
    rank=0
    if len(infra_values)>0:
        factor = 0
        for i,weight in zip(infra_values[0],[x[1] for x in col]):
            factor+=(i*weight)
        hostel=college.iloc[index]['no_of_student_hostel'] 
#         for weighted average
        factor=factor/9
        if factor+hostel is not 0:
            rank=factor+hostel
        else:
            rank=1
    else:
        rank=1
        
    Np=college.loc[index]['Phd_students']
    Nt=college.loc[index]['Total_students']
    PTR=college.loc[index]['PTR ratio']
    Pteach=college.loc[index]['Phd_teacher']
    Tfac=college.loc[index]['Total_faculty']
    Pu=college.loc[index]['Publications']
    patent=college.loc[index]['Patent']
    NG=college.loc[index]['No_gradutaes']
    OtherStud=college.loc[index]['other_state_student']
    ForeignStudent=college.loc[index]['Foreign-student']
    Placed=college.loc[index]['Students_Placed']
    
    f=0.30*((Nt*15+Np*5)*0.30 + PTR*0.40+(Pteach/Tfac)*0.30) + 0.20*(Pu*0.75 + patent*0.25) + 0.20*(Np*0.40+Pteach*0.60)+ 0.10*(rank)+ 0.10*(0.65*(OtherStud/Nt)+0.35*(ForeignStudent/Nt))+0.10*(Placed)
    college.ix[index,'Rank']=f
    
    
    
    
    
    
    
```

```python
college['Normalize_Rank']=0
```

```python
for index , row in college.iterrows():
    max=college['Rank'].max()
    min=college['Rank'].min()
    rank=college.loc[index]['Rank']
    rank=(rank-min)/(max-min)
    college.ix[index,'Normalize_Rank']=rank
    
    
    
```
